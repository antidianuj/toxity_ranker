```python
import pandas as pd
import numpy as np

df = pd.read_csv('toxity_data.csv')
df.drop(['id'], axis=1, inplace=True)
df.head()
```

```python
sentences=list(df['comment_text'].values)
labels=df[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].values


rankers=[]
classer=[]
for i in range(len(labels)):
    considered_vec=labels[i]
    if np.sum(considered_vec)==0:
        rankers.append(0.0)
        classer.append('not cyberbully')
    else:
        sub_rank=0.0
        for j in range(len(considered_vec)):
            if considered_vec[j]==1:
                sub_rank+=1.0
        rankers.append(sub_rank)
        classer.append('cyberbully')





#shuffle sentences, classer and rankers
from sklearn.utils import shuffle
sentences, rankers, classer = shuffle(sentences, rankers, classer, random_state=42)


#remove excessive non cyberbully sentences
sent=[]
rank=[]
clas=[]

for i in range(len(classer)):
    senty=sentences[i]
    #check length of senty
    if len(senty)<40:
        if classer[i]=='cyberbully':
            sent.append(sentences[i])
            rank.append(rankers[i])
            clas.append(classer[i])
        else:
            if np.random.rand()<0.1:
                sent.append(sentences[i])
                rank.append(rankers[i])
                clas.append(classer[i])


# size_of_dataset=1000
# sent=sent[:size_of_dataset]
# rank=rank[:size_of_dataset]
# clas=clas[:size_of_dataset]


#create empty dataframe
df = pd.DataFrame(columns=['sentence', 'ranking', 'class'])
df['sentence']=sent
df['ranking']=rank
df['class']=clas


#save to csv
df.to_csv('ranking_class_dataset.csv', index=False)
df.head(20)
```
